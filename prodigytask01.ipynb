```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the datasets
df=pd.read_csv("heart disease uci.csv")
```

```python
#overview
print(df.head(10))
```

```python
print("\n Shape of the data:",df.shape)
print("\n Colums of the data:",df.columns)
```

```python
categorical_vars = ["sex", "cp", "fbs", "restecg", "exang", "slope", "ca", "thal","num"]
continuous_vars = ["age", "trestbps", "chol", "thalch", "oldpeak"]

print(categorical_vars)
print(continuous_vars)
```

```python
#plotting the categorical  variable
plt.figure(figsize=(6,4))
sns.countplot(x='sex',data=df,color="red")
plt.title("Distribution of sex (cateogical variable)")
plt.xlabel("Sex ( Male, Female)")
plt.ylabel("count")
plt.show()
```

```python
# plotting the catelogical variable

plt.figure(figsize=(6,4))
sns.countplot(x='num',hue="num", data=df,palette="Reds")
plt.title("Distribution of Heart Disease (Categorical Variable)")
plt.xlabel("num")
plt.ylabel("Count")
plt.legend(
    labels=["0 = No Disease", "1 = Mild", "2 = Moderate", "3 = Severe", "4 = Very Severe"], 
    title="Severity Levels",
    loc="upper right"
)
plt.show()
```

```python
# plotting a continous variable

plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=20, kde=True,color="red")
plt.title("Distribution of Age (Continuous Variable)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#plotting the continous variable

plt.figure(figsize=(6,4))
sns.histplot(df['chol'], bins=30, kde=True, color="red")
plt.title("Distribution of Cholesterol (Continuous Variable)")
plt.xlabel("Cholesterol (mg/dl)")
plt.ylabel("Frequency")
plt.show()
```
